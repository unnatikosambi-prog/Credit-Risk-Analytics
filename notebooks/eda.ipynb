##Feature Engineering.
# Feature Engineering: Age Group
Created an `age_group` column by grouping borrowers into age ranges. This makes age-based default analysis easier and more meaningful than using individual ages.


```python
clean_data["age_group"] = pd.cut(
    clean_data["person_age"],
    bins=[18, 25, 35, 45, 60, 100],
    labels=["18-25", "26-35", "36-45", "46-60", "60+"]
)
```

```python
clean_data["age_group"].value_counts()
```

### Feature Engineering: Loan Size

Created a `loan_size` column by categorizing loan amounts into different size groups. This helps analyze whether larger loans are associated with higher default risk.

```python
clean_data["loan_size"] = pd.cut(
    clean_data["loan_amnt"],
    bins=[0, 5000, 10000, 20000, 35000],
    labels=["Small", "Medium", "Large", "Very Large"]
)
```

```python
clean_data["loan_amnt"].value_counts()
```

```python
clean_data["income_category"] = pd.qcut(
    clean_data["person_income"],
    q=4,
    labels=["Low", "Medium", "High", "Very High"]
)
```

```python
clean_data["income_category"].value_counts()
```
