## Correlation Heatmap

A correlation heatmap was generated to examine the relationships between numerical variables in the dataset. This analysis helps identify variables that move together and highlights potential factors influencing borrower behavior and loan defaults.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical columns
numeric_data =clean_data .select_dtypes(include="number")

# Calculate correlation
correlation_matrix = numeric_data.corr()

# Create the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```

### Observation

- Person age and credit history length show a strong positive correlation (0.88).
- Loan amount has a moderate positive correlation with loan percent income (0.57).
- Loan interest rate shows a moderate positive relationship with loan default status (0.32).
- Person income has a weak negative correlation with loan default status (-0.17), suggesting higher-income borrowers tend to default less frequently.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=clean_data,
    x="person_income",
    y="loan_amnt",
    hue="loan_status"
)

plt.title("Income vs Loan Amount by Loan Status")
plt.xlabel("Annual Income")
plt.ylabel("Loan Amount")
plt.show()
```

## Income Distribution by Loan Status

A box plot was created to compare the income distribution of borrowers who defaulted and those who successfully repaid their loans. This helps identify whether income levels differ between the two groups.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

sns.boxplot(
    data=clean_data,
    x="loan_status",
    y="person_income"
)

plt.title("Income Distribution by Loan Status")
plt.xlabel("Loan Status (0 = No Default, 1 = Default)")
plt.ylabel("Annual Income")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

sns.histplot(
    data=clean_data,
    x="loan_amnt",
    bins=20,
    kde=True
)

plt.title("Distribution of Loan Amount")
plt.xlabel("Loan Amount")
plt.ylabel("Number of Borrowers")

plt.show()
```
